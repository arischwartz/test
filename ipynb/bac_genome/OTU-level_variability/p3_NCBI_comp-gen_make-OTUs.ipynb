# Alignment-based QC

```python
%%bash -s "$workDir"
# dowloading mothur silva databases

cd $1

if ! [ -d mothur_silva_db ]; then
    mkdir mothur_silva_db
fi

cd mothur_silva_db

if ! [ -e $1silva_ref_aln_mothur.fasta ]; then
    curl -o silva_Bac.zip http://www.mothur.org/w/images/9/98/Silva.bacteria.zip && unzip silva_Bac.zip
    curl -o silva_Euk.zip http://www.mothur.org/w/images/1/1a/Silva.eukarya.zip && unzip silva_Euk.zip
    curl -o silva_Arc.zip http://www.mothur.org/w/images/3/3c/Silva.archaea.zip && unzip silva_Arc.zip
fi
```

```python
%%bash -s "$workDir"

cd $1'mothur_silva_db'

cat silva.bacteria/silva.bacteria.fasta \
    silva.eukarya/silva.eukarya.fasta \
    Silva.archaea/silva.archaea.fasta \
    > silva_ref_aln_mothur.fasta

printf "Number of sequences in mothur silva fasta: "
grep -c ">" silva_ref_aln_mothur.fasta
```

```python
%%bash -s "$workDir"
# filtering column positions in silva db

cd $1'mothur_silva_db'

mothur "#filter.seqs(vertical=t, \
    fasta=silva_ref_aln_mothur.fasta, \
    processors=24)" > /dev/null

printf "Number of sequences post-filter: "
grep -c ">" silva_ref_aln_mothur.filter.fasta
```

### Alignment of dataset to SILVA

```python
silvaFile = os.path.join(workDir, 'mothur_silva_db', 'silva_ref_aln_mothur.filter.fasta')
cmd = 'mothur "#align.seqs(candidate={}, template={}, processors=20, flip=T)"'.format(ampliconFile, silvaFile)
!$cmd | tail -n 30
```

```python
alnFile = os.path.splitext(ampliconFile)[0] + '.align'
cmd = 'mothur "#filter.seqs(fasta={}, vertical=t, processors=24)"'.format(alnFile)
!$cmd | tail -n 30
```

```python
alnFiltFile = os.path.splitext(alnFile)[0] + '.filter.fasta'
cmd = 'mothur "#summary.seqs(fasta={}, processors=24)"'.format(alnFiltFile)
!$cmd | tail -n 25
```

# Usearch pipeline

```python
# adding sequence 'size'; size=1 (no duplicate sequences)

alnFiltRnFile = os.path.splitext(alnFiltFile)[0] + '.rn.fna'
!perl -pe 's/[-.]+//g unless /^>/; s/(.+)/\$1;size=1/ if /^>/' \
    $alnFiltFile > $alnFiltRnFile

!head -n 4 $alnFiltRnFile
```

```python
alnFiltRnSortFile = os.path.splitext(alnFiltRnFile)[0] + '_sorted.fna'
!usearch \
    -sortbysize $alnFiltRnFile \
    -output $alnFiltRnSortFile \
    -minsize 1
    
!head -n 10 $alnFiltRnSortFile
print '----------------------'
!tail -n 10 $alnFiltRnSortFile
```

```python
otuFile = 'otus.fna'
!usearch \
    -cluster_otus $alnFiltRnSortFile \
    -otus $otuFile
```

```python
otuSnFile = os.path.splitext(otuFile)[0] + 'sn.fna'
cmd = '\'{print ">" "OTU" "." NR "::" $seq}\''
!bioawk -c fastx $cmd $otuFile | perl -pe 's/::/\n/g' > $otuSnFile
!head -n 4 $otuSnFile
```

# Mapping reads

```python
!usearch \
        -usearch_global $alnFiltRnSortFile \
        -db $otuSnFile \
        -strand plus -id 0.97 \
        -blast6out otusn_map.txt \
        -threads 20        
```

# Making an index table linking genome_name --> OTU

* genome_name -> seq_ID
* seq_ID -> ssu_ID
* ssu_ID -> OTU_ID

```python
# genome_name -> seq_ID
!grep ">" ../bac_complete_rn/*fna | \
    perl -pe 's/.+bac_complete_rn\///; s/:>/\t/' > genome_seqID.txt
!head -n 4 genome_seqID.txt
```

```python
# genome_ID -> ssu_ID
!grep ">" ../rnammer/*.fna | \
    perl -ne 'print unless /ssu_all/' | \
    perl -pe 's/.+rnammer\///; s/\.fna:>/\t/; s/ .+//' > genomeID_ssuID.txt

!wc -l genomeID_ssuID.txt  
!head -n 4 genomeID_ssuID.txt    
```

## Plotting number of taxa per OTU

```python
%%R
df.idx = read.delim('genomeID_ssuID.txt', sep='\t', header=FALSE) %>%
    mutate(V1 = gsub('_rrn$', '.fna', V1))
colnames(df.idx) = c('genome_ID', 'ssu_ID')
df.idx %>% head
```

```python
%%R -w 900 -h 300
df = read.delim('otusn_map.txt', sep='\t', header=FALSE) %>%
    mutate(V1 = gsub(';size=[0-9]+', '', V1))

df.j = inner_join(df, df.idx, c('V1' = 'ssu_ID'))

df.j %>% head
```

```python
%%R -w 900 -h 300
# check: number of OTUs per genome should be 1 (unless outlier ssu genes in genome)
df.multotu = df.j %>%
    distinct(genome_ID, V2) %>%
    group_by(genome_ID) %>%
    mutate(n=n()) %>%
    ungroup() %>%
    filter(n > 1) 

df.multotu %>% nrow %>% print
df.multotu %>% head %>% as.data.frame
```

```python
%%R -w 900 -h 300
df.j.s = df.j %>%
    group_by(V2) %>%
    summarize(n_genomes=n()) %>%
    ungroup()

df.j.s$V2 = reorder(df.j.s$V2, -df.j.s$n_genomes)

ggplot(df.j.s %>% distinct(V2), aes(V2, n_genomes)) +
    geom_point() +
    theme_bw() +
    labs(x='OTU', y='Number of genomes in OTU') +
    theme(
        axis.text.x = element_blank()
    )
```

```python
%%R -w 900 -h 300
# OTUs that are not singletons
df.j.s = df.j %>%
    group_by(V2) %>%
    summarize(n_genomes=n()) %>%
    ungroup() %>%
    filter(n_genomes > 1)

df.j.s$V2 = reorder(df.j.s$V2, -df.j.s$n_genomes)

ggplot(df.j.s %>% distinct(V2), aes(V2, n_genomes)) +
    geom_point() +
    labs(x='OTU', y='Number of genomes in OTU') +
    theme_bw() +
    theme(
        axis.text.x = element_blank()
    )
```

```python
%%R
# OTUs that are not singletons
df.ns = df.j %>%
    group_by(V2) %>%
    mutate(n_genomes=n()) %>%
    ungroup() %>%
    filter(n_genomes > 1)

# status
cat('Number of non-singleton OTUs:', df.ns %>% distinct(V2) %>% nrow, '\n')

# writing table
outFile = 'otusn_map_nonSingle.txt'
write.table(df.ns, outFile, sep='\t', quote=FALSE, row.names=FALSE, col.names=FALSE)
```
